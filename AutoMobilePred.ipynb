## Outlier Handling

```python
def outlier_remove(df):
    
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3-Q1
    
    min_bound = Q1 - 1.5*IQR
    max_bound = Q3 + 1.5*IQR
    
    df_out = df[~((df < (min_bound)) | (df > (max_bound))).any(axis=1)]
    print("Before: ",df.shape, "After: ", df_out.shape )
    
    return df_out
```

```python
outlier_remove(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression,ElasticNet
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error
import math
from sklearn.model_selection import GridSearchCV
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

## Linear Regression

```python
reg = LinearRegression()
reg.fit(X_train,y_train)
lr_pred = reg.predict(X_test)
lr_train_pred = reg.predict(X_train)
```

```python
print(f"y = {reg.intercept_} + {reg.coef_[0]}")
```

```python
print("-"*10,"Test Pred","-"*10)
print("R2: ",r2_score(y_test,lr_pred))
print("MAE: ",mean_absolute_error(y_test,lr_pred))
print("RMSE: ",math.sqrt(mean_squared_error(y_test,lr_pred)))

print("-"*10,"Train Pred","-"*10)
print("R2: ",r2_score(y_train,lr_train_pred))
print("MAE: ",mean_absolute_error(y_train,lr_train_pred))
print("RMSE: ",math.sqrt(mean_squared_error(y_train,lr_train_pred)))
```

## Random Forest
